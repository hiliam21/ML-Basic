## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
# получаем число
num5 = str(input("Введите пятизначное число:"))
# инвертируем 3 центральные цифры
inv_num5 = num5[0] + num5[3:0:-1] + num5 [4]
print(num5 + ' -> ' + inv_num5)
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
# получаем общее число дней до отпуска
days_remaining = int(input("Введите число дней до отпуска:"))
# делим на 7 и отбрасываем все float
weekends = int(days_remaining / 7 * 2)
print(str(days_remaining) + ' -> ' + str(weekends))
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
# получаем параметры
length = int(input("Введите длину:"))
width = int(input("Введите ширину:"))
size = int(input("Введите желаемый размер куска:"))


if (0 < size <= width * length  # проверяем чтобы размер куска был больше нуля и меньше чем площадь
        and (size % width == 0 or size % length == 0)): # проверяем чтобы размер делился нацело хотя бы на одну из сторон
    result = True
    print(f"{length}, {width}, {size} -> {result}")
else:
    result = False
    print(f"{length}, {width}, {size} -> {result}")
```
